```python
import presto
import pandas as pd
import numpy as np
import pulsar
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
def presto_query(query, user='test', catalog='pulsar', schema='public/default', host='localhost', port=8081):
    conn = presto.dbapi.connect(
        host=host,
        port=port,
        user=user,
        catalog=catalog,
        schema=schema)
    cur = conn.cursor()
    cur.execute(query)
    data = cur.fetchall()
    columns = cur.description
    df = pd.DataFrame(data)
    df.columns = [x[0] for x in columns]
    return df.replace('', np.nan)
```

```python
df_allocations = presto_query("SELECT * FROM allocation_topic", schema='public/default')
df_allocations.head()
```

```python
df_transactions = presto_query("SELECT * FROM transactions", schema='public/default')
df_transactions.head()
```

```python
df_payouts = presto_query("SELECT * FROM payouts", schema='public/default')
df_payouts.head()
```

```python
# make sure payoutids are unique
print(len(df_payouts), len(df_payouts['payoutid'].unique()))

print(len(df_allocations), len(df_allocations['payoutid'].unique()))
```

```python
# process df to plot
# Create traces
fig = go.Figure()

for supplier in df_transactions['supplier'].unique():
    if df_transactions.loc[df_transactions['supplier']==supplier, 'supplierbehaviorprob'].tolist()[0] == 0.0:
        line = dict(color='red', width=4, dash='dash')
        name="0.0"
    if df_transactions.loc[df_transactions['supplier']==supplier, 'supplierbehaviorprob'].tolist()[0] == 0.2:
        line = dict(color='blue', width=4, dash='dash')
        name="0.2"
    if df_transactions.loc[df_transactions['supplier']==supplier, 'supplierbehaviorprob'].tolist()[0] == 0.4:
        line = dict(color='orange', width=4, dash='dash')
        name="0.4"
    if df_transactions.loc[df_transactions['supplier']==supplier, 'supplierbehaviorprob'].tolist()[0] == 0.6:
        line = dict(color='green', width=4, dash='dash')
        name = "0.6"
    if df_transactions.loc[df_transactions['supplier']==supplier, 'supplierbehaviorprob'].tolist()[0] == 0.8:
        line = dict(color='royalblue', width=4, dash='dash')
        name="0.8"
    if df_transactions.loc[df_transactions['supplier']==supplier, 'supplierbehaviorprob'].tolist()[0] == 1.0:
        line = dict(color='purple', width=4, dash='dash')
        name = "1"
    d = df_transactions[df_transactions['user']==supplier].sort_values(by="transnum")
    fig.add_trace(go.Scatter(x=d['transnum'], y=d['balance'],
                    mode='lines',
                    name=name,
                    line=line))

fig.show()
```
