```python
import presto
import pandas as pd
import numpy as np
```

```python
class PrestoDB:
    def __init__(self, 
                 host='localhost', 
                 port=8081, 
                 user='test-user2', 
                 catalog='pulsar', 
                 schema='public/default'):
        conn = presto.dbapi.connect(
            host=host,
            port=port,
            user=user,
            catalog=catalog,
            schema=schema)
        self.cur = conn.cursor()

    def get_df(self, query):
        self.cur.execute(query)
        data = self.cur.fetchall()
        columns = self.cur.description
        df = pd.DataFrame(data)
        df.columns = [x[0] for x in columns]
        return df.replace('', np.nan)
    
    def fetch_all(self, query):
        self.cur.execute(query)
        return self.cur.fetchall()

presto_obj = PrestoDB()
```

```python
query = "SELECT * FROM generator_test WHERE firstname = 'Lucy'"
df = presto_obj.get_df(query)
df.head()
```
